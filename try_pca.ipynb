```python
from sklearn.decomposition import PCA
import pandas as pd
from pycaret.regression import * 
from sklearn.ensemble import RandomForestRegressor
```

```python
data = pd.read_csv("train.csv")
```

```python
pca_a = PCA(n_components=1)
pca_b = PCA(n_components=1)
pca_c = PCA(n_components=1)
pca_d = PCA(n_components=1)

a = ["섭씨 온도(°⁣C)", "절대 온도(K)"]
b = ["이슬점 온도(°C)", "상대 습도 (%)"]
c = ["대기압(mbar)", "포화 증기압(mbar)", "실제 증기압(mbar)", "증기압 부족량(mbar)"]
d = ["수증기 함량 (g/kg)","공기 밀도 (g/m**3)"]
```

```python
x_a = pca_a.fit_transform(data[a])
x_b = pca_b.fit_transform(data[b])
x_c = pca_c.fit_transform(data[c])
x_d = pca_d.fit_transform(data[d])
```

```python
dic = {
    "a" : x_a.flatten(),
    "b" : x_b.flatten(),
    "c" : x_c.flatten(),
    "d" : x_d.flatten(),
    "e" : data["풍향 (deg)"].values,
    "f" : data["월"].values,
    "target" : data["풍속 (m/s)"].values
}
datas = pd.DataFrame(dic)
```

```python
reg = setup(datas, train_size = 0.9,target = 'target',normalize=True,normalize_method="minmax",categorical_features=["f"])
```
